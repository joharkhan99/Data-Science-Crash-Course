```python
import pandas_datareader as pdr
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
startdate = datetime(2006, 1, 1)
enddate = datetime(2022, 1, 1)
```

```python
bank_of_america = pdr.data.DataReader("BAC", 'stooq', startdate, enddate)
citi_group = pdr.data.DataReader("CG", 'stooq', startdate, enddate)
goldman_sachs = pdr.data.DataReader("GS", 'stooq', startdate, enddate)
JP_Morgan = pdr.data.DataReader("JPC", 'stooq', startdate, enddate)
morgan_stanley = pdr.data.DataReader("MS", 'stooq', startdate, enddate)
wells_fargo = pdr.data.DataReader("WF", 'stooq', startdate, enddate)
```

```python
tickers = ['BAC','CG','GS','JPC','MS','WF']
```

```python
bank_stocks = pd.concat([bank_of_america, citi_group, goldman_sachs, JP_Morgan, morgan_stanley, wells_fargo], 
                        axis=1, keys=tickers)

bank_stocks.head()
```

```python
bank_stocks.columns.names = ['Bank Ticker', 'Stock Info']
bank_stocks.head()
```

```python
for tick in tickers:
    print(tick, bank_stocks[tick]['Close'].max())
```

```python
bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```

```python
returns = pd.DataFrame()

for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()

returns.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

import plotly
import cufflinks as cf
cf.go_offline()
```

```python
for tick in tickers:
    bank_stocks[tick]['Close'].plot(label=tick, figsize=(15, 6))
plt.legend()
```

```python
# 2nd way
bank_stocks.xs(key='Close', axis=1, level='Stock Info').plot()
```

```python
# 3rd way
bank_stocks.xs(key='Close', axis=1, level='Stock Info').iplot()
```

```python
bank_of_america['Close'].loc['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(label='30 day Moving Avg')
bank_of_america['Close'].loc['2008-01-01':'2009-01-01'].plot(label='BAC Close')
plt.legend()
```

```python
sns.heatmap(bank_stocks.xs(key='Close', axis=1, level='Stock Info').corr(), annot=True)
```
