```python
import numpy as np
x = np.arange(0, 100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.plot(x, y)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax.plot(x, y)
ax2.plot(x, y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

ax1.plot(x, z)
ax2.plot(x, y)
ax2.set_xlim([20, 22])
ax2.set_ylim([30, 50])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,2))

axes[0].plot(x, y, color='green', lw=3, ls='--')
axes[1].plot(x, z, color='maroon', lw=5)

plt.tight_layout()
plt.show()
```
