```python
import pandas as pd
```

```python
movie = pd.read_csv("movie.csv")
```

```python
movie.head()
```

```python
movie.columns
```

```python
# what we need is that movie id and title
movie = movie[['movieId', 'title']]
movie.head()
```

```python
rating = pd.read_csv("rating.csv")
```

```python
rating.head()
```

```python
rating.columns
```

```python
# what we need is that user id, movie id and rating
rating = rating[['userId', 'movieId', 'rating']]
rating.head()
```

```python
data = pd.merge(movie, rating)
```

```python
data.head(10)
```

```python
data.shape
```

```python
# lets make a pivot table in order to make rows/indexes as users and columns are movies. And values are rating
pivot_table = data.pivot_table(index=['userId'], columns=['title'], values='rating')
pivot_table.head()
```

```python
"""
Now lets make a scenario, we have movie web site and "Bad Boys (1995)" movie are watched and rated by people. The question
is that which movie do we recommend these people who watched "Bad Boys (1995)" movie.

In order to answer this question we will find similarities between "Bad Boys (1995)" movie and other movies.
"""
movie_watched = pivot_table['Bad Boys (1995)']

# find correlation between "Bad Boys (1995)" and other movies
similarity_with_other_movies = pivot_table.corrwith(movie_watched)
similarity_with_other_movies = similarity_with_other_movies.sort_values(ascending=False)
```

```python
similarity_with_other_movies.dropna(inplace=True)
```

```python
similarity_with_other_movies
```

```python
# It can be concluded that we need to recommend "Cop and ½ (1993)" movie to people 
# who watched "Bad Boys (1995)".
```
