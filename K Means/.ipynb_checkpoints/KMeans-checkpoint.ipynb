```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generate isotropic Gaussian blobs for clustering.
# basically artificial data
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
data[0].shape
```

```python
# data[0] contains the values/points
# data[1] contains the cluster number they belong to
plt.scatter(data[0][:, 0], data[0][:, 1], c=data[1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(data[0][:, 0], data[0][:, 1], c=data[1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=100, marker='s')
```

```python
# predicted labels
kmeans.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:, 0], data[0][:, 1], c=kmeans.labels_)

ax2.set_title('Original')
ax2.scatter(data[0][:, 0], data[0][:, 1], c=data[1])
```
