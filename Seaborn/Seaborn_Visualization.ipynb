```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
sns.jointplot(data=df, x='fare', y='age')
```

```python
sns.displot(data=df, x='fare', kde=True, hue='sex', bins=30)
```

```python
sns.boxplot(data=df, x='class', y='age')
```

```python
sns.swarmplot(data=df, x='class', y='age', hue='class')
```

```python
sns.countplot(data=df, x='sex')
```

```python
sns.heatmap(data=df.corr(), annot=True, fmt=".2f", cmap='coolwarm')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

sns.histplot(data=df[df.sex=='male'], x='age', ax=axes[0]).set_title("Sex: Male")
sns.histplot(data=df[df.sex=='female'], x='age', ax=axes[1]).set_title("Sex: Female")

plt.tight_layout()
plt.show()
```

```python
grid = sns.FacetGrid(data=df, col='sex')
grid.map(sns.histplot, 'age')
```

```python
grid = sns.FacetGrid(data=df[df.survived==1], col='sex')
grid.map(sns.histplot, 'age')
```

```python
sns.scatterplot(data=df[df.alive=='yes'], x='age', y='fare', hue='sex', style='sex')
```
