```python
import numpy as np
import pandas as pd

import seaborn as sns
```

```python
# Generate Fake Data
np.random.seed(0)

prices = np.random.uniform(200, 800, 1000)
feature1 = np.random.uniform(800, 1100, 1000)
feature2 = np.random.uniform(800, 1100, 1000)

data={
    'prices': prices,
    'feature1': feature1,
    'feature2': feature2
}

pd.DataFrame(data).to_csv('fake_reg.csv')
```

```python
df = pd.read_csv('fake_reg.csv', index_col=0)
```
